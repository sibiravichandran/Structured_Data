```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_parquet('/content/drive/MyDrive/Structured_Data_Assignment/train.parquet')
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
df.info()
```

```python
pat_took_td = df['Patient-Uid'][df['Incident'] == 'TARGET DRUG'].unique()
pat_took_td
```

```python
df_pat_took_td = df[df['Patient-Uid'].isin(pat_took_td)].reset_index(drop=True)
df_pat_took_td
```

```python
df_pat_took_td.sort_values(['Patient-Uid', 'Date']).reset_index(drop=True)
```

```python
# DataFrame with Incident column contains only 'TARGET DRUG'
df_pat_took_td_only = df_pat_took_td[df_pat_took_td['Incident']=='TARGET DRUG'].reset_index(drop=True)
df_pat_took_td_only
```

```python
# Data Frame with first occurence of 'TARGET DRUG' in each Patient-Uid
fir_dt_td = df_pat_took_td_only.groupby(['Patient-Uid'])['Date'].min().reset_index()
fir_dt_td
```

```python
# Data Frame with last occurence of 'TARGET DRUG' in each Patient-Uid
last_dt_td = df_pat_took_td_only.groupby(['Patient-Uid'])['Date'].max().reset_index()
last_dt_td
```

```python
# Dataframe containing both first and last dates of 'TARGET DRUG' in each Patient-Uid
df_td_lfd = fir_dt_td[['Patient-Uid', 'Date']]
```

```python
df_td_lfd.rename(columns = {'Date':'First_Date'}, inplace=True)
```

```python
df_td_lfd['Last_Date'] = last_dt_td['Date'].values
```

```python
df_td_lfd
```

```python
# DataFrame containing difference between first and last date
df_td_lfd['Diff_bet_date'] = df_td_lfd['Last_Date'].sub(df_td_lfd['First_Date'], axis = 0)
df_td_lfd
```

```python
df_td_lfd.info()
```

```python
df_td_lfd['Diff_bet_date'] = df_td_lfd['Diff_bet_date'].dt.days
```

```python
df_td_lfd.info()
```

```python
df_td_lfd
```

```python
# Taking only the data with Patient-Uid with drop-off (Diff_bet_date < 365)
df_td_ldf_lt365 = df_td_lfd[df_td_lfd['Diff_bet_date'] < 366]
```

```python
df_td_ldf_lt365.reset_index(drop=True)
```

```python
# List with all the dropped off patients
l_drof_pat = df_td_ldf_lt365['Patient-Uid'].values
l_drof_pat
```

```python
len(l_drof_pat)
```

```python
# DataFrame with Dropped Off Patients
drof_df = df_pat_took_td[df_pat_took_td['Patient-Uid'].isin(l_drof_pat)].reset_index(drop=True)
```

```python
# DataFrame with last row of each 'Patient-Uid'
last_dt_df = drof_df.groupby(['Patient-Uid'])['Date'].max().reset_index()
last_dt_df
```

```python
# Patient with only 'TARGET DRUG' DataFrame with Droppe off patients
td_drop_df = df_pat_took_td_only[df_pat_took_td_only['Patient-Uid'].isin(l_drof_pat)].reset_index(drop=True)
# DataFrame with rows of each 'Patient-Uid' who took 'TARGET DRUG for the first time'
first_dt_td_df = td_drop_df.groupby(['Patient-Uid'])['Date'].min().reset_index()
first_dt_td_df
```

```python
# DataFrame Containing both last date of each 'Patient-Uid' w.r.t 'Date' (and) date of each 'Patient-Uid'
# who took 'TARGET DRUG' for the first time from assumed drop-off list.
tgfd_dffd = first_dt_td_df[['Patient-Uid', 'Date']]
tgfd_dffd.rename(columns = {'Date':'First_Date_td'}, inplace=True)
tgfd_dffd['Last_Date_df'] = last_dt_df['Date'].values
```

```python
tgfd_dffd
```

```python
# DataFrame containing difference between first date of 'TARGET DRUG' and last date of data for each 'Patient-Uid'
tgfd_dffd['Diff_bet_date'] = tgfd_dffd['Last_Date_df'].sub(tgfd_dffd['First_Date_td'], axis = 0)
```

```python
tgfd_dffd['Diff_bet_date'] = tgfd_dffd['Diff_bet_date'].dt.days
```

```python
tgfd_dffd
```

```python
# Taking only the data with Patient-Uid with drop-off (Diff_bet_date < 365)
tgfd_dffd_lt365 = tgfd_dffd[tgfd_dffd['Diff_bet_date'] < 366]
```

```python
tgfd_dffd_lt365.reset_index(drop=True)
```

```python
# List with all the dropped off patients
l_non_drof_pat = tgfd_dffd_lt365['Patient-Uid'].values
```

```python
l_dropoff_patients = [i for i in l_drof_pat if i not in l_non_drof_pat]
```

```python
len(l_dropoff_patients)
```

```python
df_dropoff_patients = df_pat_took_td[df_pat_took_td['Patient-Uid'].isin(l_dropoff_patients)].reset_index(drop=True)
df_dropoff_patients
```

```python
total_months = (df['Date'].max() - df['Date'].min())/np.timedelta64(1, 'M')
total_months
```

```python
# Taking only the 'TARGET DRUG' of Incident for the dropped off patients and sorting it by id and date
td_of_drof_pat = df_pat_took_td_only[df_pat_took_td_only['Patient-Uid'].isin(l_dropoff_patients)].sort_values(['Patient-Uid', 'Date']).reset_index(drop=True)
```

```python
# Selecting only the last date of 'TARGET DRUG' for the dropped off patients
df_dt_drof_pat = td_of_drof_pat.groupby(['Patient-Uid'])['Date'].max().reset_index()
```

```python
# list of date of patients who dropped off
l_dt_drof_pat = df_dt_drof_pat['Date'].values
```

```python
l_dt_drof_pat
```

```python
l_m_drof_pat = ((l_dt_drof_pat - np.datetime64(df['Date'].min())).astype('timedelta64[M]')/np.timedelta64(1, 'M')).astype('int')
l_m_drof_pat
```

```python
l_c_m_drof_pat = np.bincount(l_m_drof_pat)
l_c_m_drof_pat
```
